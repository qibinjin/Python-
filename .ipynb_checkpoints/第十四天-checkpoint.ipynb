```python
import pygame, sys, random

WINDOW_HEIGHT, WINDOW_WIDTH = 768, 512

class GameWindow(object):
    def __init__(self):
        pygame.init()
        
        pygame.display.set_mode([512, 768])
        
        icon_img =  pygame.image.load('fly_fight/res/app.ico')
    
        pygame.display.set_icon(icon_img)
        
        pygame.display.set_caption('飞机大战')
        
    def __event(self):
        event_list = pygame.event.get()
        for evet in event_list:
            if event.type == pygame.QUIT:
                self.__game_over()
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_ESCAPE:
                    self.__game_over()
                if event.key == pygame.K_SPACE:
                    print('biubiubiu~~~~')
                
            pressed_list = pygame.key.get_pressed()
            if pressed_list[pygame.K_UP]:
                print('上')
            if pressed_list[pygame.K_DOWN]:
                print('下')
            if pressed_list[pygame.K_LEFT]:
                print('左')
            if pressed_list[pygame.K_RIGHT:
                print('右')
                    
                
    def __geme_over(self):
        pygame.quit()
        sys.exit()
class GameMap(object):
    def __init__(self):
        self.num = random.randint(1, 5)
        self.img1 = pygame.image.load('fly_fight/res/img_bg_level_%d.jpg' % self.num)
        self.img2 = pygame.image.load('fly_fight/res/img_bg_level_%d.jpg' % self.num)
        
        self.img1_y = -WINDOW_HEIGHT
        self.img2_y = 0
        
        self.speed = 1
    def move_down(self):
        self.img1_y += self.speed
        self.img2_y += self.speed
class HeroPlane(self):
    def __init__(self):
        self.img = pygame.image.load('fly_fight/res/hero2.png')
        self.speed = 2  
        self.img_rect = self.img.get_rect()
        self.img_rect[0] = (WINDOW_WIDTH - self.img_rect[2]) / 2
        self.img_rect[1] = WINDOW_HEIGHT - self.img_rect[1]

    
def main():
    geme_window = GameWindow()

main()
```

```python
import json
path = 'example.txt'
records = [json.loads(line) for line in open(path)]
```

```python
time_zones = [rec['tz'] for rec in records if 'tz' in rec]
```

```python
time_zones[:10]
```

```python
from collections import defaultdict,Counter
def get_counts(sequence):
    counts = defaultdict(int)
    for x in sequence:
        counts[x] += 1
    return counts
counts = get_counts(time_zones)
# print(counts)
counts['America/New_York']
```

```python
counts = Counter(time_zones)
counts.most_common(10)
```

```python
cframe = frame[frame.a.notnull()]
operating_system = np.where(cframe['a'].str.contains('Windows'),
                           'Windows', 'Not Windows')
operating_system[:5]
```

```python
by_tz_os = cframe.groupby(['tz', operating_system])
agg_counts = by_tz_os.size().unstack().fillna(0)
agg_counts[:10]
```

```python
indexer = agg_counts.sum(1).argsort()
indexer[:10]
```

```python
count_subset = agg_counts.take(indexer)[-10:]
count_subset
```

```python
count_subset.plot(kind = 'barh', stacked = True)
normed_subset = count_subset.div(count_subset.sum(1), axis = 0)
normed_subset.plot(kind = 'barh', stacked = True)
```
