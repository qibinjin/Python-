```python
class Gun(object):
    def __init__(self, model, damage):
        self.model = model
        self.damage = damage
        self.bullet_count = 0
    def __str__(self):
        return '型号:%s 伤害:%d 子弹数量:%d' % (self.model, self.damage, self.bullet_count)
    def add_bullets(self, count):
        self.bullet_count += count
        print('添加了子弹%d颗' % count)
    def shoot(self, bad_man):
        if self.bullet_count <= 0:
            print('清先添加子弹')
            return
        self.bullet_count -= 1
        if bad_man:
            bad_man.hurt(self)
class Player(object):
    def __init__(self, name, gun = None, hp = 100):
        self.name = name
        self.gun = gun
        self.hp = hp
    def __str__(self):
        if self.hp <= 0:
            return('%s[%d]已经死亡' % (self.name, self.hp))
        if not self.gun:
            return('%s[%d]没有佩戴枪' % (self.name, self.hp))
        return('%s[%d]{%s}' % (self.name, self.hp, self.gun))
    def hurt(self, gun):
        self.hp -= gun.damage
        if self.hp <= 0:
            self.hp = 0
            print('%s[%d]被%s击毙' % (self.name, self.hp, gun.model))
        else:
            print('%s[%d]被%s击中' % (self.name, self.hp, gun.model))
    def fire(self,bad_man):
        if self.gun.bullet_count<=0:
            self.gun.add_bullets(10)
        self.gun.shoot(bad_man)
            
def main():
    ak47 = Gun('AK-47', 50)
    police = Player('警察', ak47)
    print(police)
    bad_man = Player('土匪',hp = 70)
    print(bad_man)
    police.fire(bad_man)

main()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('ggplot')

x, y = np.meshgrid(np.arange(-1, 1, 0.01), np.arange(-1, 1, 0.01))

contor = np.sqrt(x ** 2 + y ** 2)

%matplotlib inline

plt.figure()
plt.imshow(contor)
plt.colorbar()
plt.grid(False)
plt.show()
```

```python
plt.style.use('ggplot')

x_min = 0.0
x_max = 1.0
y_min = 0.0
y_max = 1.0

h = .01

xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

grid = np.c_[xx.flatten(), yy.flatten()]

pred = np.zeros((grid.shape[0], 1))

for i in range(0, len(grid)):
    if i>= (grid.shape[0] / 2):
        pred[i] = 1
        
pred = pred.reshape(xx.shape, order = 'F')

plt.figure()
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.pcolormesh(xx, yy, pred)
plt.show()
```
